```python
a = int(input("Digite o número 'a': "))
b = int(input("Digite o número 'b': "))
c = int(input("Digite o número 'c': "))
lista = [a,b,c]
maior_num = 0
num_meio = 0


if a >= b and a >= c:
    maior_num = a
    lista[0] = 0
elif b >= c:
    maior_num = b
    lista[1] = 0
else:
    maior_num = c
    lista[2] = 0
for num_atual in lista:
    if num_atual > num_meio:
        num_meio = num_atual
media = (maior_num + num_meio) / 2
print(f"A média aritmética dos dois maiores valores é {media}")
```
